```python
# to implement the decision tree using different companies salaries dataset
import pandas as pd
import graphviz  # Required for visualization

data = pd.read_csv("salaries.csv")
print(data)
```

```python
inputs = data.drop('salary_more_then_100k',axis='columns')             # remove the dependent column
target = data['salary_more_then_100k']                  
inputs      # display only the independent columns
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder     # to convert categorical labels into numerical labels
company_label = LabelEncoder()                     # for new numerical label for the company
job_label = LabelEncoder()
degree_label = LabelEncoder()

inputs['company_new'] = company_label.fit_transform(inputs['company'])   # create new columns after numberical labelEncoder
inputs['job_new'] = company_label.fit_transform(inputs['job'])
inputs['degree_new'] = company_label.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis = "columns")   # old columns will be removed and numerical left only
inputs_n
```

```python
from sklearn import tree                     # for decision tree classifier or regression tree classifier
model = tree.DecisionTreeClassifier()        # instance of decision tree classifier
model.fit(inputs_n,target)                    # fit the model
model.predict([[2,2,1]])                       # prediction from the dataset, should be 0 as google employee is 0
```

```python
model.predict([[2,0,1]])         # should be 1 as business manager is 1
```
